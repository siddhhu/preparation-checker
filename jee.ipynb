```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
data=data.drop(columns='Name')
```

```python
new_data=pd.get_dummies(data)         #particular column has many sub type then split iand use 0 &1 so we can use Logistic regression easily
new_data
```

```python
column_list=new_data.columns
```

```python
features=list(set(column_list)-set(['Probablity']))
features
```

```python
y=new_data['Probablity'].values                 #storing output variabe in y
print(y)
```

```python
x=new_data[features].values                 #storing output variabe in x
print(x)
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.4,random_state=0) 
```

```python
logistic=LogisticRegression()
```

```python
logistic.fit(train_x,train_y)
```

```python
prediction=logistic.predict(test_x)       #prediction from the test data 
prediction
```

```python
confusion_matrix=confusion_matrix(test_y, prediction)     #confusion matrix   
confusion_matrix
```

```python
accuracy_score=accuracy_score(test_y, prediction)        #calculating the accuracy
print(accuracy_score)                                    #This is the accuracy it can be mor accurate if we romove some more 
                                                         
```

```python
print("No of misclassified value :",(test_y!=prediction).sum())
```
